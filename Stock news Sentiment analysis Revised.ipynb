```python
from ipywidgets import interact, interactive
import ipywidgets as widgets
```

```python
company_name=widgets.Dropdown(options=['apple','Tesla','microsoft'],value='apple',description='company name:')
sources_name=widgets.Dropdown(options=['bloomberg','business-insider'],value='bloomberg',description='sources')
domains_name=widgets.Dropdown(options=['http://www.bloomberg.com','http://www.businessinsider.com'],value='http://www.bloomberg.com',description='domains name:')
```

```python
ui=widgets.HBox((company_name,sources_name,domains_name))
```

```python
def news_options(company,sources,domains):
    all_articles = client.get_everything(q=company,
                                     sources=sources,
                                      domains=domains,
                                      from_param='2023-03-01',
                                      to='2023-03-28',
                                      language='en')
    content=[]
    date=[]
    for i in range(len(all_articles['articles'])):
        extract=all_articles['articles'][i]
        content.append(extract['content'])
        date.append(extract['publishedAt'])
    news_data=pd.DataFrame(data={'content':content,'date':date})
    ss=[]
    sid = SentimentIntensityAnalyzer()
    for sentence in news_data['content']:
        ss.append(sid.polarity_scores(str(sentence)))
    sentiment=pd.DataFrame(data=ss)
    print(sentiment.head(20))
    return sentiment
```

```python
out=widgets.interactive_output(news_options,{'company':company_name,'sources':sources_name,'domains':domains_name})
display(ui,out)
```
